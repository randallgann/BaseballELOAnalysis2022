```python
# creating a copy of the data
gamesData = df.copy()
```

```python
# convert date column to datetime format
gamesData['date'] = pd.to_datetime(gamesData['date'])
```

```python
# we have games that haven't been played yet - removing those
today = date.today().strftime("%Y-%m-%d")
gamesData = gamesData[gamesData['date']<str(today)]
```

```python
gamesData.shape
```

```python
gamesData.head()
```

```python
gamesData["fav_538_won"] = gamesData.apply(
    lambda x: 1
    if x["rating_prob1"] > x["rating_prob2"]
    and x["score1"] > x["score2"]
    or x["rating_prob1"] < x["rating_prob2"]
    and x["score1"] < x["score2"]
    else 0,
    axis=1,
)
```

```python
gamesData["fav_538_prob"] = gamesData.apply(
    lambda x: x["rating_prob1"]
    if x["rating_prob1"] > x["rating_prob2"]
    else x["rating_prob2"],
    axis=1,
)
```

```python
# checking a random sample of 20 after adding fav_538_won column
gamesData.sample(n=20, random_state=1)
```
